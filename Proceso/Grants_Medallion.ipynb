## Widgets

```python
%python
dbutils.widgets.removeAll()
```

```python
%sql
--- crear widgets

create widget text storageName default "adlssmartprojectdev13";
create widget text catalogo default "project_dev";
create widget text container default "raw";
create widget text schemaBronze default "bronze";
create widget text schemaSilver default "silver";
create widget text schemaGold default "gold";
```

```python
%python
storageName = dbutils.widgets.get("storageName")
container = dbutils.widgets.get("container")
schemaBronze = dbutils.widgets.get("schemaBronze")
schemaSilver = dbutils.widgets.get("schemaSilver")
schemaGold = dbutils.widgets.get("schemaGold")
catalogo = dbutils.widgets.get("catalogo")

# guardar el widget en una variable
```

## DEVELOPERS

```python
%sql
--- DAR PERMISOS AL CATALOGO Y SCHEMA

GRANT USE CATALOG ON CATALOG ${catalogo} TO `Developers`;
GRANT USE SCHEMA ON SCHEMA ${catalogo}.${schemaBronze} TO `Developers`;
GRANT USE SCHEMA ON SCHEMA ${catalogo}.${schemaSilver} TO `Developers`;
GRANT USE SCHEMA ON SCHEMA ${catalogo}.${schemaGold} TO `Developers`;
```

```python
%sql
--- DAR PERMISO DE LECTURA

GRANT SELECT ON SCHEMA ${catalogo}.${schemaBronze} TO `Developers`;
GRANT SELECT ON SCHEMA ${catalogo}.${schemaSilver} TO `Developers`;
GRANT SELECT ON SCHEMA ${catalogo}.${schemaGold} TO `Developers`;
```

```python
%sql
--- DAR PERMISO DE CREAR TABLAS

GRANT CREATE ON SCHEMA ${catalogo}.${schemaBronze} TO `Developers`;
GRANT CREATE ON SCHEMA ${catalogo}.${schemaSilver} TO `Developers`;
GRANT CREATE ON SCHEMA ${catalogo}.${schemaGold} TO `Developers`;
GRANT CREATE ON SCHEMA ${catalogo}.${container} TO `Developers`;
```

```python
%sql
--- dar permiso para el volumen

GRANT READ VOLUME ON VOLUME ${catalogo}.${container}.datasets TO `Developers`;

-- o si también quieres escritura:
GRANT WRITE VOLUME ON VOLUME ${catalogo}.${container}.datasets TO `Developers`;
```
